```python
df=pd.read_csv('Steel_industry_data.csv')
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(
    df['date'],
    format='%d/%m/%Y %H:%M',
    errors='coerce'
)

df['Hour'] = df['date'].dt.hour
df['Day_of_Week'] = df['date'].dt.day_name()
df['Month'] = df['date'].dt.month
df['Weekend'] = df['date'].dt.dayofweek.apply(lambda x: 1 if x >= 5 else 0)
```

```python
df['Power_Factor_Ratio'] = (
    df['Leading_Current_Power_Factor'] /
    df['Lagging_Current_Power_Factor']
)

df.head()
```

```python
threshold = df['Usage_kWh'].quantile(0.75)

df['High_Load'] = np.where(
    df['Usage_kWh'] > threshold,
    1,
    0
)

print(df['High_Load'].value_counts())
```

```python
Q1 = df['Usage_kWh'].quantile(0.25)
Q3 = df['Usage_kWh'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df['Usage_kWh'] < lower) |
    (df['Usage_kWh'] > upper)
]

print("Number of Outliers:", len(outliers))
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x=df['Usage_kWh'])
plt.title('Usage kWh Outliers')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(6,3))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr_target = numeric_df.corr()['Usage_kWh']

top_features = corr_target.abs().sort_values(
    ascending=False
)

print(top_features.head(4))
```

```python
avg_load = df.groupby(
    'Load_Type'
)['Usage_kWh'].mean()

avg_load.plot(
    kind='bar',
    figsize=(6,3)
)

plt.title('Average Energy Consumption by Load Type')
plt.ylabel('Average Usage kWh')
plt.show()
```

```python
hourly_usage = df.groupby(
    'Hour'
)['Usage_kWh'].mean()

plt.figure(figsize=(6,3))

plt.plot(
    hourly_usage.index,
    hourly_usage.values,
    marker='o'
)

plt.title('Average Energy Usage by Hour')
plt.xlabel('Hour')
plt.ylabel('Usage kWh')
plt.grid()
plt.show()
```
